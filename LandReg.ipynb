```python
from db_flask.manage import LandRegistryRecord, db
from sqlalchemy import func
from datetime import datetime
import numpy as np
import pandas as pd
from matplotlib import cm
```

```python
qry =  db.session.query(LandRegistryRecord.dst, func.avg(LandRegistryRecord.price).label('average')).filter(LandRegistryRecord.county.contains('London')).filter(LandRegistryRecord.date_of_transfer>=datetime(2010, 1, 1)).group_by(LandRegistryRecord.dst)
```

```python
avg_prices = qry.all()
```

```python
from operator import itemgetter
```

```python
p_max = max(avg_prices,key=itemgetter(1))[1]
p_min = min(avg_prices,key=itemgetter(1))[1]
```

```python
p_min
```

```python
ap = [p for a,p in avg_prices]
qnts = list(pd.Series(ap).quantile([.1, .2, .3, .4, .5, .6, .7, .8, .9]))
```

```python
def quantile_pos(price, qnts):
    try:
        return next(x[0] for x in enumerate(qnts) if x[1] > price)
    except Exception:
        return len(qnts) + 1
```

```python
quantile_pos(10000000, qnts)
```

```python
df = pd.DataFrame(avg_prices)

df['qnt_pos'] = df['average'].apply(lambda x: quantile_pos(x, qnts))
```

```python
preffix = '.district.{}'
for ix, val in df.iterrows():
    i = val['qnt_pos']/10.
    rgb = list((np.round(np.array(cm.afmhot_r(i)[:3]) * 255)).astype(int))
    pre = preffix.format(val['dst'])
    fll = '{ fill: #'+''.join(['{:02X}'.format(c) for c in rgb]) + '; }'
    print(pre, fll)
```

```python
qrya =  db.session.query(LandRegistryRecord.area, func.avg(LandRegistryRecord.price).label('average')).filter(LandRegistryRecord.county.contains('London')).filter(LandRegistryRecord.date_of_transfer>=datetime(2010, 1, 1)).group_by(LandRegistryRecord.area)
```

```python
avg_prices_area = qrya.all()
apa = [p for a,p in avg_prices_area]
qntsa = list(pd.Series(apa).quantile([.1, .2, .3, .4, .5, .6, .7, .8, .9]))
```

```python
dfa = pd.DataFrame(avg_prices_area)
dfa['qnt_pos'] = dfa['average'].apply(lambda x: quantile_pos(x, qntsa))
```

```python
preffix = '.area.{}'
for ix, val in dfa.iterrows():
    i = val['qnt_pos']/10.
    rgb = list((np.round(np.array(cm.afmhot_r(i)[:3]) * 255)).astype(int))
    pre = preffix.format(val['area'])
    fll = '{ fill: #'+''.join(['{:02X}'.format(c) for c in rgb]) + '; }'
    print(pre, fll)
```
